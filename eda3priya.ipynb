```python
# 11. Find Average Value of Reviews
# Check unique problematic values
data['Reviews'].unique()

# Replace '3.0M' with '3.0'
data['Reviews'] = data['Reviews'].replace('3.0M', '3.0')

# Convert column into float
data['Reviews'] = data['Reviews'].astype('float')

# Find average reviews
avg_reviews = data['Reviews'].mean()

print("Average Reviews:", avg_reviews)
```
